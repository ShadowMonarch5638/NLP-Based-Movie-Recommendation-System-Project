AST(Abstract Syntax Trees)

The ast module in Python helps analyze and understand the structure of Python code. It's useful for tasks like checking code quality, finding errors, and making automated changes to code. It's like a tool that helps Python programs understand and work with other Python code.


> Indented block



```python
# Import the 'ast' (Abstract Syntax Trees) module for parsing and working with Python literals
import ast
```

```python
def convert(obj):
    # Initialize an empty list to store the extracted names
    L = []
    # Iterate through the list of dictionaries in 'obj'
    for i in ast.literal_eval(obj):
        # Append the 'name' field of each dictionary to the list 'L'
        L.append(i['name'])

    return L

def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
      if counter != 3:
            L.append(i['name'])
            counter +=1
      else:
          break
      return L
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies_df['genres']= movies_df['genres'].apply(convert)
movies_df['keywords']=movies_df['keywords'].apply(convert)

movies_df['cast']= movies_df['cast'].apply(convert3)
movies = movies_df  # Replace with your data source

# Apply the fetch_director function to the 'crew' column
movies['crew'] = movies['crew'].apply(fetch_director)

```

```python
# Apply lambda functions to remove spaces from elements within the 'genres', 'keywords', and 'crew' columns in the 'movies' DataFrame.
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

COUNTVECTORIZER:

.CountVectorizer is converts collection of the text document to matrix
 of token count this produce the sparse representation

.CountVectorizer is capture how often the word is occured in the document it count frequency of word and it show in matrix formats





```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
# arranges the data into a matrix.
# Each row stands for a document each column for a word, and the values show the word counts in each document.
new_df['tags'].fillna('', inplace=True)

vectors=cv.fit_transform(new_df['tags']).toarray()
```

COSINE SIMILARITY

Cosine similarity is used because it's good at measuring the similarity between items, particularly in text data. It's scale-invariant, angle-based, efficient, and works well for recommendation systems, clustering, and classification tasks.

```python
# Import the 'cosine_similarity' function from the 'sklearn.metrics.pairwise' module
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calculate the cosine similarity between vectors in the 'vectors' matrix
cosine_similarity_matrix = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
similarity[0]
```

```python
# Sort the similarity scores for the first vector (index 0) and return the indices of the top 5 most similar vectors.

# 1. Use the 'enumerate' function to associate each similarity score with its index.
# 2. Sort the enumerated similarity scores in descending order (highest similarity first).
# 3. Extract the indices (vector indices) of the top 5 similar vectors.

similar_indices = sorted(enumerate(similarity[0]), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), key=lambda x: x[1], reverse=True)[1:6]


    for i in movies_list:
        recommended_movie_title = new_df.iloc[i[0]]['title']  # Access the 'title' from the DataFrame
        print(recommended_movie_title)

```
